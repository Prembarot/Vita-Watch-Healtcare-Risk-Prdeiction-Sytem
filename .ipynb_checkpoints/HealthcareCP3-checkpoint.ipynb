```python
# Step 1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 2

# Load the dataset
data = pd.read_csv('C:/capstone/archive/health_monitoring_system.csv')

# Display the first few rows of the dataset
print(data.head())

# Check for null values
print(data.isnull().sum())

# Drop rows with missing values
data = data.dropna()

# Summary of the dataset
print(data.describe())

# Visualize the distribution of the 'Type' column
sns.countplot(x='Type', data=data)
plt.title('Distribution of Health Conditions')
plt.show()
```

```python
# Step 3

# Creating new feature columns
data['Age_Group'] = pd.cut(data['Temperature'], bins=[0, 20, 40, 60, 80, 100], labels=['Very Low', 'Low', 'Moderate', 'High', 'Very High'])

# Features and target
X = data[['Temperature', 'Heart Rate', 'Pulse', 'BPSYS', 'BPDIA', 'Respiratory Rate', 'Oxygen Saturation', 'PH']]
y = data['Type']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 4 and 5

# Creating a RandomForestClassifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Training the model
model.fit(X_train, y_train)

# Predicting the test set results
y_pred = model.predict(X_test)

# Evaluate model performance
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Adjust threshold for a more sensitive prediction
y_pred_prob = model.predict_proba(X_test)[:, 1]
threshold = 0.1  # Lowering threshold to increase sensitivity
y_pred_new = np.where(y_pred_prob > threshold, 1, 0)

print(classification_report(y_test, y_pred_new))
```

```python
# Step 6

# Feature importance
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh', color='skyblue')
plt.title('Feature Importance')
plt.show()

# Pairplot for initial data exploration
sns.pairplot(data, hue='Type', palette='husl')
plt.show()

# Interactive 3D Scatter Plot
fig = px.scatter_3d(data, x='Temperature', y='Heart Rate', z='BPSYS', color='Type', symbol='Age_Group',
                    title='3D Scatter Plot of Temperature, Heart Rate, and Systolic Blood Pressure')
fig.show()

# Boxplot for Age Groups vs Temperature
plt.figure(figsize=(12,6))
sns.boxplot(x='Age_Group', y='Temperature', data=data, palette='Set3')
plt.title('Boxplot of Age Groups vs Temperature')
plt.show()
```

```python
# Step 7

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Initialize the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Healthcare Monitoring Dashboard"),
    
    dcc.Dropdown(
        id='feature-dropdown',
        options=[
            {'label': 'Temperature', 'value': 'Temperature'},
            {'label': 'Heart Rate', 'value': 'Heart Rate'},
            {'label': 'Pulse', 'value': 'Pulse'},
            {'label': 'BPSYS', 'value': 'BPSYS'},
            {'label': 'BPDIA', 'value': 'BPDIA'},
            {'label': 'Respiratory Rate', 'value': 'Respiratory Rate'},
            {'label': 'Oxygen Saturation', 'value': 'Oxygen Saturation'},
            {'label': 'PH', 'value': 'PH'},
            {'label': 'Age Group', 'value': 'Age_Group'}
        ],
        value='Temperature'
    ),
    dcc.Graph(id='feature-importance')
])

@app.callback(
    Output('feature-importance', 'figure'),
    [Input('feature-dropdown', 'value')]
)
def update_graph(selected_feature):
    if selected_feature in data.columns:
        fig = px.histogram(data, x=selected_feature, color='Type', barmode='group',
                           title=f'Histogram of {selected_feature} by Type')
    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8061)  # Change the port number to 8061
```

```python
# Step 8

import numpy as np
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Function to monitor and alert based on new data with adjusted threshold
def monitor_and_alert(new_data):
    new_data = np.array(new_data).reshape(1, -1)
    prediction_prob = model.predict_proba(new_data)[:, 1]
    threshold = 0.1  # Lowered threshold to increase sensitivity further
    prediction = np.where(prediction_prob > threshold, 1, 0)
    if prediction == 1:
        return "ALERT: High risk detected!"
    else:
        return "All good. No high risk detected."

# Example usage with new patient data (for Dash callback)
def generate_alert(temp, hr, pulse, bpsys, bpdia, resp_rate, o2_sat, ph):
    new_patient_data = [temp, hr, pulse, bpsys, bpdia, resp_rate, o2_sat, ph]
    return monitor_and_alert(new_patient_data)

# Initialize the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Healthcare Monitoring Dashboard"),
    
    dcc.Dropdown(
        id='feature-dropdown',
        options=[
            {'label': 'Temperature', 'value': 'Temperature'},
            {'label': 'Heart Rate', 'value': 'Heart Rate'},
            {'label': 'Pulse', 'value': 'Pulse'},
            {'label': 'BPSYS', 'value': 'BPSYS'},
            {'label': 'BPDIA', 'value': 'BPDIA'},
            {'label': 'Respiratory Rate', 'value': 'Respiratory Rate'},
            {'label': 'Oxygen Saturation', 'value': 'Oxygen Saturation'},
            {'label': 'PH', 'value': 'PH'},
            {'label': 'Age Group', 'value': 'Age_Group'}
        ],
        value='Temperature'
    ),
    dcc.Graph(id='feature-importance'),
    
    html.H3("Alert System"),
    dcc.Input(id='input-temp', type='number', placeholder='Temperature'),
    dcc.Input(id='input-hr', type='number', placeholder='Heart Rate'),
    dcc.Input(id='input-pulse', type='number', placeholder='Pulse'),
    dcc.Input(id='input-bpsys', type='number', placeholder='BPSYS'),
    dcc.Input(id='input-bpdia', type='number', placeholder='BPDIA'),
    dcc.Input(id='input-resp', type='number', placeholder='Respiratory Rate'),
    dcc.Input(id='input-o2sat', type='number', placeholder='Oxygen Saturation'),
    dcc.Input(id='input-ph', type='number', placeholder='PH'),
    html.Button('Check Alert', id='check-alert', n_clicks=0),
    html.Div(id='alert-output')
])

@app.callback(
    Output('feature-importance', 'figure'),
    [Input('feature-dropdown', 'value')]
)
def update_graph(selected_feature):
    if selected_feature in data.columns:
        fig = px.histogram(data, x=selected_feature, color='Type', barmode='group',
                           title=f'Histogram of {selected_feature} by Type')
    return fig

@app.callback(
    Output('alert-output', 'children'),
    [Input('check-alert', 'n_clicks')],
    [dash.dependencies.State('input-temp', 'value'),
     dash.dependencies.State('input-hr', 'value'),
     dash.dependencies.State('input-pulse', 'value'),
     dash.dependencies.State('input-bpsys', 'value'),
     dash.dependencies.State('input-bpdia', 'value'),
     dash.dependencies.State('input-resp', 'value'),
     dash.dependencies.State('input-o2sat', 'value'),
     dash.dependencies.State('input-ph', 'value')]
)
def update_alert(n_clicks, temp, hr, pulse, bpsys, bpdia, resp_rate, o2_sat, ph):
    if None in [temp, hr, pulse, bpsys, bpdia, resp_rate, o2_sat, ph]:
        return "Please fill in all values."
    return generate_alert(temp, hr, pulse, bpsys, bpdia, resp_rate, o2_sat, ph)

if __name__ == '__main__':
    app.run_server(debug=True, port=8061)  # Change the port number to 8061
```
